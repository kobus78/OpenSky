```python
import urllib.request
import json
from pandas.io.json import json_normalize
import threading, time, signal
import pandas as pd
import datetime
```

```python
def capture_df():
    with urllib.request.urlopen("https://opensky-network.org/api/states/all") as response:
        raw = response.read()  
    data = json.loads(raw)
    #a_state_vector = data['states'][10]; print(len(a_state_vector))
    #print( len(data['states']) )        
    df_tmp = json_normalize(data, 'states'); print(len(df_tmp))
    #df_tmp['time'] = data['time']
    #dt = pd.to_datetime(data['time'], unit='ns', origin=datetime.datetime.now().date())
    #dt = pd.to_datetime(data['time'], unit='s', origin=datetime.datetime.now().date())
    raw_time = data['time']
    date_time = pd.to_datetime(raw_time, unit='s', origin='unix')
    #df_tmp['datetime'] = date_time
    df_tmp['rawtime'] = raw_time
    return(raw_time,date_time,df_tmp)
```

## Construct raw dataframe

```python
df = pd.DataFrame()
for i in range(10):
    print(i)
    with urllib.request.urlopen("https://opensky-network.org/api/states/all") as response:
        raw = response.read()
    df_current = capture_df()
    df = df.append(df_current)
    time.sleep(50)
#df
```

```python
from datetime import datetime
import os

from apscheduler.schedulers.blocking import BlockingScheduler

def do_it():
    print('Tick! The time is: %s' % datetime.now())
    with urllib.request.urlopen("https://opensky-network.org/api/states/all") as response:
        raw = response.read()
    _,date_time,df = capture_df()
    fname = '_'.join(f'{date_time}'.split())
    df.to_csv(f'data/{fname}.csv', header=False)

if __name__ == '__main__':
    scheduler = BlockingScheduler()
    scheduler.add_executor('processpool')
    scheduler.add_job(do_it, 'interval', seconds=60)
    print('Press Ctrl+{0} to exit'.format('Break' if os.name == 'nt' else 'C'))

    try:
        scheduler.start()
    except (KeyboardInterrupt, SystemExit):
        pass
```
