```python
import pandas as pd
import gc
#gc.collect()
```

```python
filename_base = 'allraw-12h-apr23'
filename_in = f'{filename_base}.csv'
filename_out = f'{filename_base}-combined.csv'
```

```python
#df = pd.read_csv('allraw.csv', header=None); print(df.shape, df.size)
df = pd.read_csv(filename_in, 
                 header=None,
                 low_memory=False
                 #nrows=4000000
                ); print(df.shape, df.size)
```

```python
df.head()
```

```python
df.columns = ['idx','icao24','callsign','origin_country','time_position','last_contact','longitude','latitude',
              'baro_altitude','on_ground','velocity','true_track','vertical_rate','sensors','geo_altitude','squawk',
              'spi','position_source','rawtime']
```

```python
df.head()
```

```python
df['datetime'] = pd.to_datetime(df.rawtime, unit='s', origin='unix')
df.head()
```

```python
#df
```

```python
df.set_index('datetime', drop=True, inplace=True)
df.drop(['idx','rawtime'], axis=1, inplace=True)
df.head()
```

```python
df.sort_index(inplace=True)
```

```python
df.to_csv(filename_out)
```
